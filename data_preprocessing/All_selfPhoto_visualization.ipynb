```python
import numpy as np
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
self_photo= pd.read_csv("self_photo_addressOnly.csv")

geo_local = Nominatim(user_agent='South Korea')
```

```python
# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
    
```

```python
latitude = []
longitude =[]

for i in self_photo['도로명주소']:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
for i in range(len(latitude)):
    print(latitude[i], longitude[i], "\n")
```

```python
self_photo['위도'] = 0
self_photo['경도'] = 0
```

```python
for i in range(len(latitude)):
    self_photo['위도'][i] = latitude[i]
    self_photo['경도'][i] = longitude[i]
```
